# Warshall algorithm to find reachability between all pairs of vertices in a graph.
Problem Description:
Understand transitive closure of a graph. Use Python to compute reachability using matrix method.

#Theory
Warshall algorithm finds whether a path exists between every pair of vertices. It updates the adjacency matrix step by step. If there is a path from i to j through k, then the entry becomes 1.

## Algorithm
1. Represent graph using adjacency matrix.
2. For each intermediate vertex k
3. Check all pairs i and j
4. Update value if path exists through k
4. Repeat for all vertices


```python
INF = float('inf')

def warshall(graph):
    n = len(graph)

    for k in range(n):
        for i in range(n):
            for j in range(n):
                graph[i][j] = graph[i][j] or (graph[i][k] and graph[k][j])

    return graph

def floyd(graph):
    n = len(graph)

    for k in range(n):
        for i in range(n):
            for j in range(n):
                graph[i][j] = min(graph[i][j], graph[i][k] + graph[k][j])

    return graph

n = 5
graph = []

print("Enter adjacency matrix (0 or 1) for 5 vertices:")

for i in range(n):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    graph.append(row)

warshall_graph = [row[:] for row in graph]

floyd_graph = []

for i in range(n):
    row = []
    for j in range(n):
        if i == j:
            row.append(0)
        elif graph[i][j] == 1:
            row.append(1)
        else:
            row.append(INF)
    floyd_graph.append(row)

warshall_result = warshall(warshall_graph)
floyd_result = floyd(floyd_graph)

print("\nTransitive Closure Matrix (Warshall):")
for row in warshall_result:
    print(row)

print("\nShortest Path Matrix (Floyd):")
for row in floyd_result:
    print(row)

print("\nPath Existence:")
for i in range(n):
    for j in range(n):
        if warshall_result[i][j]:
            print(f"Path exists from {i} to {j}")
        else:
            print(f"No path from {i} to {j}")

print("\nComparison:")
print("Warshall Algorithm → Checks reachability (Yes/No path).")
print("Floyd Algorithm → Finds shortest path distances.")
```

## Find shortest paths between all pairs of vertices using Floyd Warshall algorithm.

#Understand all pairs shortest path problem. Implement dynamic programming approach in Python.

* Given a weighted graph with positive or negative edge weights, find
shortest distances between every pair of vertices.

# Theory
Floyd Warshall algorithm computes shortest paths between all pairs of nodes. It updates the distance matrix by checking if a shorter path exists through an intermediate vertex. Works with negative weights but fails with negative cycles.

Algorithm
1. Represent graph using distance matrix.
2. Initialize matrix with given weights.
3. Set distance of node to itself as 0.
4. For each intermediate vertex k
5. Update distance[i][j] = min(distance[i][j], distance[i][k] + distance[k][j])
6. Repeat for all vertices

```python
INF = float('inf')

def floyd_warshall(graph):
    n = len(graph)
    dist = [row[:] for row in graph]
    next_node = [[None if graph[i][j] == INF else j for j in range(n)] for i in range(n)]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
                    next_node[i][j] = next_node[i][k]

    return dist, next_node


def detect_negative_cycle(dist):
    for i in range(len(dist)):
        if dist[i][i] < 0:
            return True
    return False


def get_path(u, v, next_node):
    if next_node[u][v] is None:
        return []
    path = [u]
    while u != v:
        u = next_node[u][v]
        path.append(u)
    return path


n = int(input("Enter number of vertices: "))

graph = []
print("Enter adjacency matrix row-wise (use INF for no edge):")

for i in range(n):
    row = input().split()
    cleaned_row = []

    for x in row:
        if x == "INF":
            cleaned_row.append(INF)
        else:
            cleaned_row.append(int(x))

    graph.append(cleaned_row)

dist, next_node = floyd_warshall(graph)

print("\nShortest Distance Matrix:")
for row in dist:
    print(row)

if detect_negative_cycle(dist):
    print("\nNegative cycle detected")
else:
    print("\nNo negative cycle")

u = int(input("\nEnter source vertex: "))
v = int(input("Enter destination vertex: "))

path = get_path(u, v, next_node)

if path:
    print("Shortest path:", path)
    print("Distance:", dist[u][v])
else:
    print("No path exists")
```

```python
INF = float('inf')

graph = [
[0, 3, INF, 7],
[8, 0, 2, INF],
[5, INF, 0, 1],
[2, INF, INF, 0]
]

result = floyd_warshall(graph)

print("Shortest distance matrix:")

for row in result:
    print(row)
```

```python
INF = float('inf')

def floyd_with_path(graph):
    n = len(graph)
    dist = [[graph[i][j] for j in range(n)] for i in range(n)]
    next_node = [[None if graph[i][j] == INF else j for j in range(n)] for i in range(n)]
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
                    next_node[i][j] = next_node[i][k]
    return dist, next_node

def construct_path(u, v, next_node):
    if next_node[u][v] is None:
        return []
    path = [u]
    while u != v:
        u = next_node[u][v]
        path.append(u)
    return path

import heapq
def dijkstra(graph, src):
    n = len(graph)
    dist = [INF] * n
    dist[src] = 0
    pq = [(0, src)]
    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v in range(n):
            if graph[u][v] != INF and dist[v] > dist[u] + graph[u][v]:
                dist[v] = dist[u] + graph[u][v]
                heapq.heappush(pq, (dist[v], v))
    return dist

n = int(input("Enter number of vertices: "))
graph = []
print("Enter adjacency matrix (use INF for no edge):")
for i in range(n):
    row = input().split()
    row = [INF if x == "INF" else int(x) for x in row]
    graph.append(row)

dist, next_node = floyd_with_path(graph)
negative_cycle = any(dist[i][i] < 0 for i in range(n))
print("\nFloyd–Warshall Result:")
for row in dist:
    print(["∞" if val == INF else val for val in row])
print("\nNegative cycle detected" if negative_cycle else "\nNo negative cycle detected")

src, dest = map(int, input("\nEnter source and destination vertices: ").split())
if dist[src][dest] == INF:
    print(f"No path from {src} to {dest}")
else:
    path = construct_path(src, dest, next_node)
    print(f"Shortest path from {src} to {dest}: {path} with distance {dist[src][dest]}")

src_dijkstra = int(input("\nEnter source vertex for Dijkstra: "))
dijkstra_dist = dijkstra(graph, src_dijkstra)
print("\nDijkstra distances from source:")
print(["∞" if val == INF else val for val in dijkstra_dist])
print("\nFloyd–Warshall distances from source:")
print(["∞" if val == INF else val for val in dist[src_dijkstra]])
```
